### (b) IF/ELSE statements

- Create two lists: 

- ` list_fruits = ["apple","orange","pear","banana"]`
- ` list_colors = ["blue","red","green","orange"]`

- Create a new variable ` message = "orange"`
- Create an if/else that tests whether ` message ` is a fruit and a color.  Print a message if the test is true.

<br>

- **Hint**: Use the `in` command to test whether an element is part of a list (see Lecture 4) 

```python
# Write your answer here:
list_fruits = ["apple", "orange", "pear", "banana"]
list_colors = ["blue", "red", "green","orange"]

message = "orange"

if (message in list_fruits) & (message in list_colors): 
    print("Message is a fruit and a color")
elif (message in list_colors):
    print("Message is only a color")
elif (message in list_fruits):
    print("Message is only a fruit")
else:
    print("Message is neither a fruit or a color")
```

### (f) Run loop through different datasets and compute statistics

- Create a list with the `.csv` dataset names (including their relative subfolder `data_quiz/`)

    - Read the corresponding `.csv` dataset.
    - Print the mean of the variable `percent_ages0to14` in each dataset

Note: This exercise tells you how much the proportion of the population that is between 0 and 14 has varied from year to year using real data from the World Development Index.



```python
# Write your answer here:
quiz = pd.read_csv('data_quiz/wdi_ageprop_2020.csv')
list_varnames = ["percent_ages0to14", "percent_ages15to64"]


for quiz in list_varnames:
    data = pd.read_csv(quiz)
    mean_value = data['percent_ages0to14'].mean()
    print(f"The mean of 'percent_ages0to14' in {quiz} is: {mean_value}")



```
